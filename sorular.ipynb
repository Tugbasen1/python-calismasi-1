### Soru 24: KDV Hesaplama
KDV hesaplama programı yazınız:
- Temel gıda: %1

- Kitap: %8
- Temizlik: %18
- Elektronik: %20
- Diğer: %18

```python
# Ürün fiyatı
fiyat = float(input("Ürün fiyatını giriniz: "))

# Ürün tipi seçimi
print("\nÜrün Tipini Seçiniz:")
print("1- Temel Gıda (%1)")
print("2- Diğer Gıda (%10)")
print("3- Kitap (%8)")
print("4- Temizlik (%18)")
print("5- Elektronik (%20)")
print("6- Diğer (%18)")

secim = input("Seçiminiz (1-6): ")

# KDV oranı belirleme
if secim == "1":
    kdv_orani = 1
elif secim == "2":
    kdv_orani = 10
elif secim == "3":
    kdv_orani = 8
elif secim == "4":
    kdv_orani = 18
elif secim == "5":
    kdv_orani = 20
elif secim == "6":
    kdv_orani = 18
else:
    print("Geçersiz seçim yaptınız.")
    exit()

# Hesaplama
kdv_tutari = fiyat * kdv_orani / 100
kdvli_fiyat = fiyat + kdv_tutari

print("\nKDV Oranı: %", kdv_orani)
print("KDV Tutarı:", kdv_tutari, "TL")
print("KDV Dahil Fiyat:", kdvli_fiyat, "TL")

```

### Soru 25: Taksi Ücreti Hesaplama
Bir taksi yolculuğu ücretini hesaplayan program yazınız:
- Açılış: 12₺
- km başı: 8₺
- Gece 00.00-06.00 arası: %20 zamlı
- Hafta sonu: %10 zamlı
- 50 km üzeri: %15 indirimli

```python
# Sabitler
acilis_ucreti = 12
km_basi = 8

# Kullanıcıdan bilgiler
km = float(input("Gidilen kilometreyi giriniz: "))
saat = int(input("Saat kaç? (0-23): "))
gun = input("Gün nedir? (haftaici/haftasonu): ").lower()

# Temel ücret hesaplama
ucret = acilis_ucreti + (km * km_basi)

# Gece zammı (00-06 arası)
if 0 <= saat < 6:
    ucret *= 1.20  # %20 zam

# Hafta sonu zammı
if gun == "haftasonu":
    ucret *= 1.10  # %10 zam

# 50 km üzeri indirim
if km > 50:
    ucret *= 0.85  # %15 indirim

print("Toplam ödenecek ücret:", round(ucret, 2), "TL")

```

### Soru 26: ATM Para Çekme
ATM'den para çekme simülasyonu programı yazınız:
- Hesap bakiyesi: 5000₺
- Çekilmek istenen miktarın 100'ün katı olması kontrolü
- Bakiyeden büyük olmaması kontrolü
- Günlük limiti (3000₺) aşmaması kontrolü

```python
bakiye = 5000
gunluk_limit = 3000

print("Mevcut bakiyeniz:", bakiye, "TL")

cekilecek_tutar = int(input("Çekmek istediğiniz tutarı giriniz: "))

# 100'ün katı kontrolü
if cekilecek_tutar % 100 != 0:
    print("Hata: Çekilecek tutar 100'ün katı olmalıdır.")

# Günlük limit kontrolü
elif cekilecek_tutar > gunluk_limit:
    print("Hata: Günlük 3000 TL limitini aşamazsınız.")

# Bakiye kontrolü
elif cekilecek_tutar > bakiye:
    print("Hata: Yetersiz bakiye.")

# Başarılı işlem
else:
    bakiye -= cekilecek_tutar
    print("İşlem başarılı!")
    print("Çekilen tutar:", cekilecek_tutar, "TL")
    print("Kalan bakiye:", bakiye, "TL")

```

### Soru 27: Maaş Hesaplama
Bir iş yerinde çalışanın maaşını hesaplayan program yazınız:
- Saatlik ücret: 100₺
- Haftalık 45 saate kadar: normal ücret
- 45-55 saat arası: 1.5 katı
- 55 saat üzeri: 2 katı
- Haftada 70 saatten fazla çalıştırılamaz

```python
saatlik_ucret = 100

calisma_saati = float(input("Haftalık çalışma saatinizi giriniz: "))

if calisma_saati > 70:
    print("Hata: Haftada 70 saatten fazla çalışılamaz.")

elif calisma_saati <= 45:
    maas = calisma_saati * saatlik_ucret

elif calisma_saati <= 55:
    normal = 45 * saatlik_ucret
    fazla = (calisma_saati - 45) * (saatlik_ucret * 1.5)
    maas = normal + fazla

else:  # 55 - 70 arası
    normal = 45 * saatlik_ucret
    fazla1 = 10 * (saatlik_ucret * 1.5)   # 45-55 arası
    fazla2 = (calisma_saati - 55) * (saatlik_ucret * 2)
    maas = normal + fazla1 + fazla2

if calisma_saati <= 70:
    print("Toplam maaş:", maas, "TL")

```

### Soru 28: Gelişmiş Not Hesaplama
Not hesaplama sistemi programı yazınız:
- Vize: %30
- Ödev: %20
- Final: %50
- Geçme notu: 60
- Final notu 45'in altındaysa direkt kalır
- Devamsızlık 10 günü geçerse devamsızlıktan kalır

```python
vize = float(input("Vize notunu giriniz: "))
final = float(input("Final notunu giriniz: "))
odev = float(input("Ödev notunu giriniz: "))
devamsizlik = int(input("Devamsızlık gün sayısını giriniz: "))

# Devamsızlık kontrolü
if devamsizlik > 10:
    print("Sonuç: Devamsızlıktan kaldınız.")

# Final baraj kontrolü
elif final < 45:
    print("Sonuç: Final notu 45 altı olduğu için kaldınız.")

else:
    ortalama = (vize * 0.30) + (final * 0.50) + (odev * 0.20)
    print("Genel ortalama:", round(ortalama, 2))
    
    if ortalama >= 60:
        print("Sonuç: Dersi geçtiniz.")
    else:
        print("Sonuç: Dersten kaldınız.")

```

- Latte: 40₺

```python
# Kahve fiyatları
print("Kahve Otomatına Hoşgeldiniz!")
print("1- Türk Kahvesi (25 TL)")
print("2- Espresso (30 TL)")
print("3- Americano (35 TL)")
print("4- Latte (40 TL)")

secim = input("Seçiminizi yapınız (1-4): ")

# Fiyat belirleme
if secim == "1":
    fiyat = 25
elif secim == "2":
    fiyat = 30
elif secim == "3":
    fiyat = 35
elif secim == "4":
    fiyat = 40
else:
    print("Geçersiz seçim!")
    exit()

# Para girişi
para = int(input("Para miktarını giriniz (5,10,20,50): "))

# Geçerli para kontrolü
if para not in [5, 10, 20, 50]:
    print("Hata: Sadece 5, 10, 20 veya 50 TL kabul edilir!")

elif para < fiyat:
    print("Yetersiz bakiye! Eksik:", fiyat - para, "TL")

else:
    ust_para = para - fiyat
    print("Kahveniz hazırlanıyor ☕")
    print("Üst para:", ust_para, "TL")
```

### Soru 30: Vücut Kitle İndeksi
Vücut Kitle İndeksi hesaplama programı yazınız: Kilo/(boy*boy) - boy metre cinsinden.
Sonuca göre:
- 18.5 altı: Zayıf
- 18.5-24.9: Normal
- 25-29.9: Fazla kilolu
- 30-34.9: 1. derece obez
- 35-39.9: 2. derece obez
- 40 üstü: 3. derece obez

```python
kilo = float(input("Kilonuzu giriniz (kg): "))
boy = float(input("Boyunuzu giriniz (metre): "))

vki = kilo / (boy * boy)

print("Vücut Kitle İndeksi:", round(vki, 2))

if vki < 18.5:
    print("Durum: Zayıf")

elif vki <= 24.9:
    print("Durum: Normal")

elif vki <= 29.9:
    print("Durum: Fazla Kilolu")

elif vki <= 34.9:
    print("Durum: 1. Derece Obez")

elif vki <= 39.9:
    print("Durum: 2. Derece Obez")

else:
    print("Durum: 3. Derece Obez")
```
